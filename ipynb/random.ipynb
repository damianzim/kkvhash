```python
import sys
sys.path.insert(0, '..')

from python.kkvhash import kkv_hash

buffer = [bytes([i]) for i in range(256)]
```

```python
from random import randint

len_min = 6
len_max = 12

count = 1000000

def get_rand() -> bytes:
    return buffer[randint(0, 255)]

def run_random() -> list:
    return [
        kkv_hash(b''.join([get_rand() for _ in range(randint(len_min, len_max))]))
        for i
        in range(count)
    ]

def run_composite_test(n: int) -> None:
    assert 0 < n < 100
    results = [
        count - len(set(run_random()))
        for i
        in range(n)
    ]
    print(results)
    print(f"The worst one: {max(results)}")
```

```python
run_composite_test(10)
```
