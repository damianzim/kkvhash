```python
# Calculate similarity

modes = {
    Mode.m1mln: '1 mln passwords',
    Mode.m1mln_key: '1 mln key',
}

plt.title('Similarity of strings for the same hash')
plt.xlabel('Similarity')
plt.ylabel('Occurrence')
plt.grid(True)

for mode, label in modes.items():
    analyzer = Analyzer(mode)
    loaded = analyzer.load()
    print(f"{label} - loaded: {loaded}")
    if loaded:
        duplicates_similarity = analyzer.calc_similarity()
        if duplicates_similarity is not None:
            similarity, occurrence = zip(*duplicates_similarity)
            plt.plot(similarity, occurrence, label=label)
            
plt.legend()
plt.show()
```
