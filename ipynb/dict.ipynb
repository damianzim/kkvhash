```python
letters = """
a	A 4 @
b	B
c	C
d	D
e	E 3
f	F
g	G
h	H
i	I
j	J
k	K
l	L
m	M
n	N
o	O 0
p	P
q	Q g
r	R
s	S 5 $
t	T
u	U
v	V u
w	W
x	X
y	Y
z	Z
"""

def gen_letters(mode: str = 'c++') -> None:

    if mode == 'python':
        print('letters =  {')
    for line in letters.strip().splitlines():
        key, values = line.split('\t')

        if mode == 'c++':
        # c++ std::map<char, std::vector<char>> (inside)
            print("{ '%s', { %s } }," % (key, ", ".join(map(lambda x: f"'{x}'", values.split()))))
        elif mode == 'python':
            # python dict
            print(f"\t'%s': [%s]," % (key, ", ".join(map(lambda x: f"'{x}'", values.split()))))
    if mode == 'python':
        print('}')
        
gen_letters('python')
```
